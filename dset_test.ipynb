```python
from tdc.multi_pred import DTI
import pandas as pd
data = DTI(name = 'DAVIS')
split = data.get_split()
print(split['train'].columns)
a1 = pd.unique(split['train']['Target_ID'])
a3 = pd.unique(split['train']['Target'])
```

```python
def get_id_seq_mapping(dataset):
    data = DTI(name=dataset)
    split = data.get_split()
    unique_target_pairs = pd.unique(split['train'][['Target_ID', 'Target']].apply(tuple, axis=1))

    # Map the unique target sequences to their corresponding UniProt IDs
    id_to_sequence = {seq: target_id for target_id, seq in unique_target_pairs}
    
    return id_to_sequence
```

```python
import os

def save_array_chunks_to_text(array, chunk_size, output_dir):
    """Save chunks of an array to text files in the specified directory."""
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    num_chunks = (len(array) + chunk_size - 1) // chunk_size  # Calculate number of chunks needed
    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = (i + 1) * chunk_size
        chunk = array[start_idx:end_idx]

        # Write chunk to a text file
        output_file = os.path.join(output_dir, f"seqs_{i+1}.txt")
        with open(output_file, 'w') as file:
            for item in chunk:
                file.write(str(item) + '\n')

def main():
    # Example array
    example_array = a3  # Replace with your array
    
    chunk_size = 100
    output_dir = "davis_seqs"  # Directory to save chunked text files
    
    save_array_chunks_to_text(example_array, chunk_size, output_dir)
    print(f"Array chunks saved to {output_dir}")

if __name__ == "__main__":
    main()
```
